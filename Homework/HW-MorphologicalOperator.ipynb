```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('finger.jpg',0)
kernel = np.ones((3,3),np.uint8)
dst = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

erosion = cv2.erode(dst,kernel,iterations = 1)
opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
dilation = cv2.dilate(opening,kernel,iterations = 1)
closing = cv2.morphologyEx(dilation, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(10, 10))
plt.subplot(231),plt.imshow(dst),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(erosion),plt.title('Erosion')
plt.subplot(233),plt.imshow(opening),plt.title('Opening')
plt.subplot(234),plt.imshow(dilation),plt.title('Dilation')
plt.subplot(235),plt.imshow(closing),plt.title('Closing')
plt.imshow(dst)
plt.show()
```
